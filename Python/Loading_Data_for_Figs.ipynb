```python
import os.path
import time
import numpy as np
from scipy import rand
from scipy.io import loadmat
%pylab inline
from __future__ import division
from __future__ import print_function
from pylab import *
import os
import time
from scipy.io import savemat, loadmat
from numpy import *
import pandas as pd
from scipy.io import loadmat
from datatank_py.DTDataFile import DTDataFile
from datatank_py.DTProgress import DTProgress
from datatank_py.DTSeries import DTSeriesGroup

def loadData(fn):  
    if dataSet == '3D_1.5um':
        fileName = '/Users/gracemclaughlin/Dropbox/Gladfelter_Gems_project/Axial_Projections/\
GEMS-grace_fps_test (axialDensityData)/3D_1.5um/'
    if dataSet == '3D_1um':
        fileName = '/Users/gracemclaughlin/Dropbox/Gladfelter_Gems_project/Axial_Projections/\
GEMS-grace_fps_test (axialDensityData)/3D_1um/'
    if dataSet == 'GEMS-grace_dropletsAndNuclei':
        fileName = '/Users/gracemclaughlin/Dropbox/Gladfelter_Gems_project/Axial_Projections/\
GEMS-grace_dropletsAndNuclei (axialDensityData)/6_7_19_nucleiMarked/gems_/'
    if dataSet == 'GEMS-grace_new2_WT':
        fileName = '/Users/gracemclaughlin/Dropbox/Gladfelter_Gems_project/Axial_Projections/\
GEMS-grace_new2 (axialDensityData)/'
    if dataSet == 'gems_no_nuc':
        fileName = '/Users/gracemclaughlin/Dropbox/Gladfelter_Gems_project/Axial_Projections/\
GEMS-grace_new (axialDensityData)/4_18_19/{0}/'.format(dataSet)     
    ############    
    hyphaName = '{0}-hyphaAxisDensity.dtbin'.format(fn) 
    fileName+=hyphaName
    ############
    with DTDataFile(fileName, readonly=True) as sfile:
        print(sfile.variable_names())
        densityKeysU = [name for name in sfile.variable_names() if 'density' in name]
        Nvids = len(densityKeysU)
        densityKeys = [[name for name in densityKeysU if '_{0}_'.format(n) in name][0] 
                       for n in np.arange(Nvids)]
        densitiesRaw = []
        for name in densityKeys:
            temp = array(sfile[name]).squeeze()
            if len(temp.shape) < 2:
                temp = temp.reshape(-1,1)
                assert len(temp.shape) == 2
            densitiesRaw.append(temp)
        arclengthKeysU = [name for name in sfile.variable_names() if 'arcLength' in name]
        arclengthKeys = [[name for name in arclengthKeysU if '_{0}_'.format(n) in name][0] 
                         for n in np.arange(Nvids)]
        arclengthRaw = [array(sfile[name]).squeeze() for name in arclengthKeys]
        ########################
        diffusivityKeysU = [name for name in sfile.variable_names() if 'diffusivity' in name]
        Nvids = len(diffusivityKeysU)
        diffusivityKeys = [[name for name in diffusivityKeysU if '_{0}_'.format(n) in name][0] 
                       for n in np.arange(Nvids)]
        diffusivityRaw = []
        for name in diffusivityKeys:
            temp = array(sfile[name]).squeeze()
            if len(temp.shape) < 2:
                temp = temp.reshape(-1,1)
                assert len(temp.shape) == 2
            diffusivityRaw.append(temp)
        #########################
        minProjectionKeysU = [name for name in sfile.variable_names() if 'minProjection' in 
                              name]
        Nvids = len(minProjectionKeysU)
        minProjectionKeys = [[name for name in minProjectionKeysU if '_{0}_'.format(n) in name][0] 
                       for n in np.arange(Nvids)]
        minProjectionRaw = []
        for name in minProjectionKeys:
            temp = array(sfile[name]).squeeze()
            if len(temp.shape) < 2:
                temp = temp.reshape(-1,1)
                assert len(temp.shape) == 2
            minProjectionRaw.append(temp)               
        #########################
        nucMedAxisKeysU = [name for name in sfile.variable_names() if 
                           'nucleiMedialAxisPosition' in name]
        Nvids = len(nucMedAxisKeysU)
        nucMedAxisKeys = [[name for name in nucMedAxisKeysU if '_{0}_'.format(n) in name][0] 
                       for n in np.arange(Nvids)]
        nucMedAxisRaw = [array(sfile[name]).squeeze() for name in nucMedAxisKeys]
        #########################
        axisPositionKeysU = [name for name in sfile.variable_names() if 'axisPosition' in name]
        Nvids = len(axisPositionKeysU)
        axisPositionKeys = [[name for name in axisPositionKeysU if '_{0}_'.format(n) 
                             in name][0] 
                       for n in np.arange(Nvids)]
        axisPositionRaw = [array(sfile[name]).squeeze() for name in axisPositionKeys]
        ##########################
        radiusKeysU = [name for name in sfile.variable_names() if 'radius' in name]
        Nvids = len(radiusKeysU)
        radiusKeys = [[name for name in radiusKeysU if '_{0}_'.format(n) in name][0] 
                       for n in np.arange(Nvids)]
        radiusRaw = []
        for name in radiusKeys:
            temp = array(sfile[name]).squeeze()
            if len(temp.shape) < 2:
                temp = temp.reshape(-1,1)
                assert len(temp.shape) == 2
            radiusRaw.append(temp)
##########################################minProjection
    densities = [-d[~isnan(d)].reshape(1,-1, d.shape[1],1).T for d in densitiesRaw]
    nanDensities = [-d.reshape(1,-1, d.shape[1],1).T for d in densitiesRaw]    
    diffusivity = []
    for d in diffusivityRaw:
        d[isnan(d)] = 0 
        diffusivity.append(d.T)
    minProjection = []
    for d in minProjectionRaw:
        d[isnan(d)] = 0 
        minProjection.append(d.T)
    arcLengths = [-d[~isnan(d)] for d in arclengthRaw]
    nucMedAxis = nucMedAxisRaw
    axisPosition = axisPositionRaw
    path = '../'
    savePath = path
    return densities,arcLengths,diffusivity,axisPosition,radiusRaw,minProjection,nucMedAxis
###################
# dataSet = 'gems_no_nuc' #4/18
# vids = ['1x_gems_noNuc001','1x_gems_noNuc003','1x_gems_noNuc005','1x_gems_noNuc007',
#         '1x_gems_noNuc009','1x_gems_noNuc011','1x_gems_noNuc013','1x_gems_noNuc015',
#         '1x_gems_noNuc017','1x_gems_noNuc019']
##################
# dataSet = 'GEMS-grace_new2_WT'
# vids = ['gems_wtNoNuc001','gems_wtNoNuc003','gems_wtNoNuc005','gems_wtNoNuc007'
#        ,'gems_wtNoNuc009','gems_wtNoNuc011','gems_wtNoNuc013','gems_wtNoNuc015'
#        ,'gems_wtNoNuc017','gems_wtNoNuc019','gems_wtNoNuc021','gems_wtNoNuc023'
#        ,'gems_wtNoNuc025','gems_wtNoNuc027','gems_wtNoNuc029','gems_wtNoNuc031'
#        ,'gems_wtNoNuc033']
##################
dataSet = 'GEMS-grace_dropletsAndNuclei'
vids = ['gems_908.2026','gems_908.2030','gems_908.2034','gems_908.3017']
##################
# dataSet = '3D_1.5um'
# vids = ['1.5stack_003','wt_gems_1.5stack003']
##################
# dataSet = '3D_1um'
# vids = ['wt-gems_1frame_1umstack005','wt-gems_1frame_1umstack007','wt-gems_1frame_1umstack007']
##################
nest = {}
for n in vids:
    print('------------------')
    print('vid:', n)
    print('------------------')
    nest[n] = {}
    densities,arcLengths,diffusivity,axisPosition,radiusRaw,minProjection,nucMedAxis = loadData(n)
    nest[n]['densities'] = densities
    nest[n]['arcLength'] = arcLengths
    nest[n]['diffusivity'] = diffusivity
    nest[n]['axisPosition'] = axisPosition
    nest[n]['radius'] = radiusRaw
    nest[n]['minProjection'] = minProjection
    nest[n]['nucleiMedialAxisPosition'] = nucMedAxis
# np.save('axialData-{0}.npy'.format(dataSet), [nest])
np.save('axialDataLinkerModified-{0}.npy'.format(dataSet), [nest])
```
